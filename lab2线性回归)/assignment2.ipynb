<span style="color:purple">1) 使用pandas库的read_csv()函数(可以参考[pandas的官方文档](https://pandas.pydata.org/docs/reference/api/pandas.read_csv.html))将训练数据集'train.csv'和测试数据集'test.csv'载入到Dataframe对象中。</span>

```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
#读取数据集
train_frame = pd.read_csv('train.csv')
test_frame = pd.read_csv('test.csv')


#转化成numpy矩阵
train = np.array(train_frame)
test = np.array(test_frame)
print(test_frame)

```

<span style="color:purple">2) 假设模型为一元线性回归模型$\hat{y}=wx+b$, 损失函数为$l(w,b)=\frac{1}{2}\sum_{i=1}^m(\hat{y}^{(i)}-y^{(i)})^2$, 其中$\hat{y}^{(i)}$表示第$i$个样本的预测值，$y^{(i)}$表示第$i$个样本的实际标签值, $m$为训练集中样本的个数。求出使得损失函数最小化的参数$w$和$b$。</span>

方法① 

将$l(w,b)$分别对$w$和$b$求导，得到
$$
\frac{\partial l(w,b)}{\partial w}=w\sum_{i=1}^m x_i^2 -\sum_{i=1}^m (y_i-b)x_i,
$$
$$
\frac{\partial l(w,b)}{\partial b}=mb -\sum_{i=1}^m (y_i-wx_i),
$$
令上述两式为零即可得到$w$和$b$的解析解：
$$
w=\frac{\sum_{i=1}^m y_i (x_i-\bar{x})}{\sum_{i=1}^m x_i^2-\frac{1}{m}(\sum_{i=1}^m x_i)^2},
$$
$$
b=\frac{1}{m}\sum_{i=1}^m(y_i-wx_i),
$$
其中$\bar{x}=\frac{1}{m}\sum_{i=1}^m x_i$为$x$的均值。



方法② 梯度下降法。手动实现梯度下降法(不使用机器学习框架，如PyTorch、TensorFlow等)来进行模型的训练。算法步骤如下：1.初始化模型参数$w$和$b$的值；2.在负梯度的方向上更新参数(批量梯度下降、小批量随机梯度下降或者随机梯度下降均可)，并不断迭代这一步骤，更新公式(以小批量随机梯度下降为例)可以写成：$$w\gets w-\frac{\eta}{\left|B\right|}\sum_{i\in{B}}x^{(i)}(wx^{(i)}+b-y^{(i)})$$, 和$$b\gets b-\frac{\eta}{\left|B\right|}\sum_{i\in{B}}(wx^{(i)}+b-y^{(i)})$$， 其中$\eta$表示学习率,$B$表示每次迭代中随机抽样的小批量，$\left|B\right|$则表示$B$中的样本数量。3. 终止条件为迭代次数达到某一上限或者参数更新的幅度小于某个阈值。

```python
# 最小二乘法
print(train.shape)
x_train = train[:,0]
x_train_square = np.square(x_train)
y_train = train[:,1]
x_test = test[:,0]
y_test = test[:,1]

x_train_ave = x_train.mean()
print(x_train_ave.shape)
print(x_train_ave)
m = x_train_square.shape[0]
print(m)

w = (y_train*(x_train-x_train_ave)).sum()/(x_train_square.sum()-x_train.sum()**2/m)
b = (y_train-w*x_train).sum()/m

print('w:',w)
print('b',b)
fig,axs = plt.subplots(sharex=True)

axs.scatter(x_test,y_test,label='origin')
axs.scatter(x_test,w*x_test+b,label='predict')
axs.legend()
plt.show()
```

```python

# 梯度下降法，批量梯度下降


# 设置初始参数
w = 0
b = 0
learning_rate = 0.1
min_update = 0.0001

# 设置一些变量
x_test = test[:,0]
y_test = test[:,1]
iter = 500

y_plot_train = []
y_plot_test = []

# 画图
plt.rcParams['font.family'] = 'Microsoft YaHei'
fig,axs = plt.subplots(1,2,sharex=True,figsize=(20,6))
axs[0].scatter(x_test,y_test,label='origin')
axs[1].scatter(x_train,y_train,label='origin')

# 迭代过程
while(iter>0):
    axs[0].scatter(x_test,w*x_test+b,label='迭代次数：'+str(16-iter))
    axs[1].scatter(x_train,w*x_train+b,label='迭代次数：'+str(16-iter))
    iter= iter-1
    
    grad_w = (x_train*(x_train*w + b - y_train)).sum()/m
    grad_b = (x_train*w + b - y_train).sum()/m
    w = w-learning_rate*grad_w
    b = b-learning_rate*grad_b
    if(learning_rate*grad_w<min_update or learning_rate*grad_b<min_update):
        print('stop')
        # break
    
    y_plot_train.append(np.square(y_train-(w*x_train+b)).sum()/y_train.shape[0])
    y_plot_test.append(np.square(y_test-(w*x_test+b)).sum()/y_test.shape[0])
    

# 画迭代的散点图
# axs[0].legend()
# axs[1].legend()
# axs[0].set_title('test数据集')
# axs[1].set_title('train数据集')
# plt.show()


# 误差曲线
fig,axs = plt.subplots(1,2,sharex=True,figsize=(20,6))
axs[0].plot(range(500),y_plot_train,label = 'train')
axs[0].legend()
axs[1].plot(range(500),y_plot_test,label = 'test')
axs[1].legend()
plt.show()

# 散点图
fig,axs = plt.subplots(1,2,sharex=True,figsize=(20,6))
axs[0].scatter(x_test,y_test,label='origin')
axs[0].scatter(x_test,w*x_test+b,label='predict')
axs[0].legend()
axs[0].set_title('test-data')
axs[1].scatter(x_train,y_train,label='origin')
axs[1].scatter(x_train,w*x_train+b,label='predict')
axs[1].legend()
axs[1].set_title('train-data')
axs[0].set_title('test数据集')
axs[1].set_title('train数据集')
plt.show()
    
    
print(w,b)
```

方法③ 

用矩阵表示，假设数据集有$m$个样本，特征有$n$维$。X=\left[ \begin{matrix} x_{11} & x_{12} & \cdots & x_{1n} & 1 \\
                         x_{21} & x_{22} & \cdots & x_{2n} & 1 \\
                         \vdots & \vdots &      & \vdots & \vdots \\
                         x_{m1} & x_{m2} & \cdots & x_{mn} & 1 \end{matrix} \right]$,
        实际标签$Y=\left[ \begin{matrix} y_{1} \\
                         y_{2} \\
                         \vdots \\
                         y_{m}\end{matrix} \right]$,
        参数$B=\left[ \begin{matrix} w_{1} \\
                         w_{2} \\
                         \vdots \\
                         w_{n} \\
                         b\end{matrix} \right]$，则解析解为$B^*=(X^T X)^{-1}X^T Y$。推导过程可参考[这篇文章](https://zhuanlan.zhihu.com/p/74157986)。

<span style="color:purple">3) 使用求解出来的线性回归模型对测试数据集'test.csv'进行预测，输出可视化结果（比如用seaborn或者matplotlib等可视化库来画出测试数据的散点图以及训练好的模型函数图像）。</span>

```python
# 求解
X = np.vstack((x_train,np.ones(x_train.shape[0]))).T
# print(X[:2],)
Y = y_train
B = np.linalg.inv(np.transpose(X)@X)@np.transpose(X)@Y
print(B)
w = B[0]
b = B[1]

# 画图
fig,axs = plt.subplots(1,2,sharex=True,figsize=(20,6))
axs[0].scatter(x_test,y_test,label='origin')
axs[0].scatter(x_test,w*x_test+b,label='predict')
axs[0].legend()
axs[0].set_title('test-data')
axs[1].scatter(x_train,y_train,label='origin')
axs[1].scatter(x_train,w*x_train+b,label='predict')
axs[1].legend()
axs[1].set_title('train-data')
plt.show()
```
